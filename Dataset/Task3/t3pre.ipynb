```python
from py3dbp import Packer, Bin, Item

packer = Packer()
# (35, 23, 13), (37, 26, 13), (38, 26, 13), (40, 28, 16),
#(42, 30, 18), (42, 30, 40), (52, 40, 17), (54, 45, 36).
max=1000000
min=1
packer.add_bin(Bin('bin1', 35, 23, 13,max))
packer.add_bin(Bin('bin2', 37, 26, 13,max))
packer.add_bin(Bin('bin3', 38, 26, 13,max))
packer.add_bin(Bin('bin4', 40, 28, 16,max))
packer.add_bin(Bin('bin5', 42, 30, 18,max))
packer.add_bin(Bin('bin6', 42, 30, 40,max))
packer.add_bin(Bin('bin7', 52, 40, 17,max))
packer.add_bin(Bin('bin8', 54, 45, 36,max))
```

```python

for i in range(0,5):
    tmp=df[df['sta_code']==kf[i]]
    for j in range(0,len(tmp)):
        l=tmp.iloc[j]['长(CM)']
        w=tmp.iloc[j]['宽(CM)']
        h=tmp.iloc[j]['高(CM)']
        q=tmp.iloc[j]['qty']
        name=tmp.iloc[j]['sku_code']
        packer.add_item(Item(name, l, w, h,min))
packer.pack()
```

```python


for bin in packer.bins:
    print(f"Bin: {bin.name}")
    print("Fitted Items:")
    for item in bin.items:
        print(f"-> {item.name}")
    print("Unfitted Items:")
    for item in bin.unfitted_items:
        print(f"-> {item.name}\n")
```
